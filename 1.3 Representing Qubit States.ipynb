## 1. Classical vs Quantum Bits

### 1.1 Statevectors(状態ベクトル)

量子物理学では、状態ベクトルを使用してシステムの状態を記述します。トラックに沿った車の位置を記述したいとします。
これは古典的なシステムなので、xを使用できます。

代わりに、"Statevector(状態ベクトル)"と呼ばれるベクトルで数のコレクションを使用することもできます。
状態ベクトルの各要素には、特定の場所で自動車を見つける確率が含まれています。

これは位置に限定されず、車が持つ可能性のある全ての速度と、車が持つ可能性のある全ての色の状態ベクトルを保持することもできます。
古典的なシステム(上記の自動車の例のような)では、本当に1つの数値しか必要としないときに巨大なベクトルを維持する必要があるため、適さないです。
ただし、この章で説明するように、状態ベクトルは、量子コンピュータを含む量子システムを追跡する非常に優れた方法なのです。

### 1.2 Qubit Notation 

古典的なビットは常に、完全に、明確に定義された状態を持っています。
それらは計算中のすべての時点で、0または1のどちらかです。
これ以上、ビットの状態に追加できる情報はありません。
従って、古典的なビット(c)の状態を書き留めるには、これら2つのバイナリ値を使用します。
例えば

c = 0

この制限は、qubitでは解除されています。qubitから0と1のどちらを取得するかは、出力を抽出するための測定が行われるときにのみ明確に定義する必要があります。
その時点で、これらの2つのオプションのいずれかにコミットする必要があります。それ以外の場合は常に、その状態は単純なバイナリ値で取得できるよりも複雑なものになります。

これらの説明方法を確認するには、最初に2つの最も単純なケースに焦点を当てます。前のセクションで見たように、測定したときに必ず結果が0になるqubit(初期化されたqubit)を準備することができます。

この状態に名前が必要です。単純に0としてしまいましょう。
同様に、1を確実に出力するqubit状態が存在します。これを1と呼ぶことにします。
これら2つの状態は完全に相互に排他的です。
qubitは絶対に0か1のどちらかを出力します。重複はありません。
これを数学で表す1つの方法は、2つの直交ベクトルを使用することです。

これは、一度にすべての情報を取り込むための表記方法です。
最初に見慣れない|と>の説明をしましょう。
彼らの仕事は、基本的に0と1のラベルがついたqubit状態を表すベクトルについて話していることを思い出させることです。
これはビット値0と1または数値0と1のようなものから区別するのに役立ちます。
Diracによって導入された"bra-ket(ブラケット表記)"と言います。

もし、ベクトルに慣れていない場合は、基本的に、特定のルールを使って操作する数値のリストと考えることができます。
高校の物理学のクラスでやるベクトルに精通している場合は、これらの規則により、ベクトルが大きさと方向で数量を記述するのに非常に適していることがわかります。例えば、オブジェクトの速度は完全にベクトルで記述されます。
ただし、量子状態にベクトルを使用する方法は、これとは少し異なります。
なので、あなたの直感に強く囚われないでください。
新しいことをする時が来ました！！

ベクトルを使用すると、|0>と|1>だけでなく、より複雑な状態を記述できます。例えば、ベクトルを考えてみましょう。

この状態が何を意味するのかを理解するには、ベクトルを操作するための数学的なルールを使用する必要があります。
具体的には、ベクトルを一緒に追加する方法と、ベクトルをスカラー倍する方法を理解する必要があります。

|0>と|1>が正規直交基底なので、これら2つの状態の組み合わせで任意の2Dベクトルを表すことができます。

この|q0>というベクトルはqubitの状態ベクトルと呼ばれ、私たちが知ることができる全てを教えてくれます。現時点では、状態ベクトルのこの特定の例について、いくつかの簡単な結論を出すことができます。
完全に|0>ではなく、完全に|1>でもないということ。
代わりに2つの線形結合によって表されます。量子力学では、通常"Superposition(重ね合わせ)"という単語を使用して、このような線形結合を記述します。

この例の状態|q0>は|0>と|1>の重ね合わせとして表現できますが、それは明確に定義されたqubit状態です。これを確認するために、qubitの操作方法を探ることができます。

### 1.3 Exploring Qubits with Qiskit

まずは、必要なツールをインポートしましょう。

```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram, plot_bloch_vector
from math import sqrt, pi
```

Qiskitでは、QuantumCircuitオブジェクトを使用して回路を格納します。
これは基本的に、回路内の量子ゲートとそれらが適用されるqubitのリストです。

```python
qc = QuantumCircuit(1) # Create a quantum circuit with one qubit
```

この量子回路では、qubitはいつも|0>状態から始まります。
"initialize()"メソッドを使用して、このqubitを任意の状態に変換できます。
initialize()に必要なベクトルをリストの形式で指定し、この状態で初期化するqubit(s)を指定します。

```python
qc = QuantumCircuit(1) # Create a quantum circuit with one qubit
initial_state = [0,1]   # Define initial_state as |1>
qc.initialize(initial_state, 0) # Apply initialisation operation to the 0th qubit
qc.draw()              # Let's view our circuit
```

次に、Qiskitのシミュレータの1つを使用して、qubitの結果の状態を表示できます。まず、statevectorシミュレータを使用しますが、様々なシミュレータとその使用法については後で説明します。

```python
backend = Aer.get_backend('statevector_simulator') # Tell Qiskit how to simulate our circuit
```

量子回路から結果を取得するために、"execute"を使用して回路を実行し、回路とバックエンドを引数として指定します。
次に、".result()"を使用してこの結果を取得します。

```python
qc = QuantumCircuit(1) # Create a quantum circuit with one qubit
initial_state = [0,1]   # Define initial_state as |1>
qc.initialize(initial_state, 0) # Apply initialisation operation to the 0th qubit
result = execute(qc,backend).result() # Do the simulation, returning the result
```

結果から、".get_statevector()"を使用して最終的な状態ベクトルを取得できます。

```python
qc = QuantumCircuit(1) # Create a quantum circuit with one qubit
initial_state = [0,1]   # Define initial_state as |1>
qc.initialize(initial_state, 0) # Apply initialisation operation to the 0th qubit
result = execute(qc,backend).result() # Do the simulation, returning the result
out_state = result.get_statevector()
print(out_state) # Display the output state vector
```

注：Pythonでは、jを使用してi(複素数)を表します。
0. + 0.j = 0
1. + 1.j = 1 

実際の量子コンピュータと同じようにqubitを測定して、結果を見てみましょう。

記号 "<" 及び "|" は<ψ|が行ベクトルであることを教えてくれています。
量子力学では、列ベクトルを"kets(ケット)"、行ベクトルを"bras(ブラ)"と呼びます。
この2つがブラケット表記を構成しています。
すべてのケット|a>には対応するブラ<a|が存在します。

#### 3 Global Phase

状態1>を測定すると、出力1が確実に得られることがわかっています。
しかし、次のような状態を書き留めることもできます。

これがどのように動作するかを確認するために、測定ルールを適用します。

ここで、複素数の大きさをとると、iの因数が消えることがわかります。
この効果は、測定された状態|x>から完全に独立しています。
従って、どの測定を検討しているかは関係ありません。
状態i|1>の確率は、|1>の確率と同じです。
測定はqubitから情報を抽出する唯一の方法であるため、これらの2つの状態は物理的に関連するすべての点で同等であることを意味します。

より一般的には、|y| = 1 の状態の全ての因子yを"global phase"と呼びます。
global phaseのみが異なる状態は、物理的に区別することはできません。

これは、"relative phase(相対位相)"として知られる、重ね合わせにおけるtermsとは異なることに注意してください。
これは、様々なタイプの測定を考えた後に、重要となります。

#### 4 The Observer Effect

振幅には特定の状態のqubitを見つける確率に関する情報が含まれていますが、qubitを測定すると、qubitの状態が確実にわかります。
例えば、次のようになります。

測定した結果|0|の状態であれば、もう一度測定しても、|0>の状態でqubitを見つける確率は100%になる。
これは、測定することで、qubitの状態が変化することを意味しています。

これをqubitの状態の崩壊と呼ぶことがあります。これは強力な効果であり、賢く使う必要があります。
例えば、計算の各ポイントでそれらの値を追跡するために各quubitを絶えず測定していたら、それらは常に|0>または|1>のいずれかの明確な状態になります。
それらは古典的なビットと何ら変わりがなく、私たちの計算は古典的な計算に簡単に置き換えることができます。
しかし、本当の量子計算を実現するには、qubitがより複雑な状態を探索できるようにする必要があります。従って、測定値は、出力を抽出する必要がある場合にのみ使用されます。これは、多くの場合、全ての測定を量子回路の最後に配置することを意味します。

Qiskitのstatevector simulatorを使用して、これを実証できます。
superposition状態でqubitを初期化しましょう。

```python
qc = QuantumCircuit(1) # Redefine qc
initial_state = [0.+1.j/sqrt(2),1/sqrt(2)+0.j]
qc.initialize(initial_state, 0)
qc.draw()
```

これによって、qubitが次の状態で初期化されているはずです。

これはシミュレータを使用して確認することができます。

```python
state = execute(qc, backend).result().get_statevector()
print("Qubit State = " + str(state))
```

これで、qubitが[0.+0.70710678j 0.70710678+0.j]という、我々が予想した状態で初期化されていることがわかりました。

それでは、このqubitを測定しましょう。

```python
qc.measure_all()
qc.draw()
```

この回路全体をシミュレーションすると、振幅の1つが常に0であることがわかります。

```python
state = execute(qc, backend).result().get_statevector()
print("State of Measured Qubit = " + str(state))
```

このセルを数回再実行して、qubitを再初期化し、再度測定することができます。
どちらの結果も同じになる可能性がありますが、qubitの状態が|0>と|1>の重ね合わせになることはありません。
興味深いことに、|0>状態のグローバルフェーズは存続しますが、これはグローバル位相であるため、実際の量子コンピュータでは測定できません。

#### 量子シミュレータに関する注意
